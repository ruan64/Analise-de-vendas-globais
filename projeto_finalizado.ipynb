2.IMPOTAÇÃO DAS BIBLIOTECAS

```python
#Impotação das Bibliotecas Pandas, Plotly e IPython
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import display, HTML
```

3.AMAZENANDO DADOS NA VARIAVEL

>



```python
#Importando a Tabela para a Variavel Por camimho
arquivo = '/content/drive/MyDrive/dataset.xlsx'
df = pd.read_excel(arquivo)

```

4.DEFININDO COLUNAS

```python
df.columns = [col.strip().lower() for col in df.columns]
```

5.FUNÇÃO ENCONTRAR COLUNAS

```python
#Criando a função encontra colunas relevantes
def encontrar_coluna(possiveis):
    for nome in possiveis:
        for c in df.columns:
            if nome in c:
                return c
    return None

col_pais = encontrar_coluna(['pais', 'country'])
col_categoria = encontrar_coluna(['categoria', 'category'])
col_subcategoria = encontrar_coluna(['sub', 'subcategoria', 'sub-category'])
col_vendas = encontrar_coluna(['venda', 'valor', 'sales', 'amount', 'total'])
col_data = encontrar_coluna(['data', 'date'])
col_desconto = encontrar_coluna(['desc', 'discount'])
col_pedido = encontrar_coluna(['pedido', 'order', 'id_pedido', 'order id'])
```

6.CONVERTER VALORES NULL EM COERCE

```python
#Retirar valores em branco
if col_data:
    df[col_data] = pd.to_datetime(df[col_data], errors='coerce')
    df['ano'] = df[col_data].dt.year
```

7.TITULO PARA O DESHBOARD

```python
#Formatando o Titulo Do Dashboard
display(HTML("<h2 style='text-align:center;'> DASHBOARD DE ANÁLISE DE VENDAS GLOBAIS</h2>"))
```

8.CRIAÇÃO DO CARTÃO TOTAL VENDAS

```python
#Somar o Total de vendas
if col_vendas:
    total = df[col_vendas].sum()
    fig1 = go.Figure(go.Indicator(
        mode="number",
        value=total,
#Titulo Do Gráfico
        title={"text": "<b>Total de Vendas</b>"},
        number={"prefix": "R$", "valueformat": ",.2f"}
    ))
#Chamado Do Gráfico
    fig1.update_layout(template="plotly_dark", height=250)
    fig1.show()
```

9.CRIAÇÃO DO GRAFICO DE PIZZA

```python
#Agrupando Categoria Por pedido
if col_categoria and col_pedido:
    grp = df.groupby(col_categoria, as_index=False)[col_pedido].count()
    if not grp.empty:
#Configurações Do Gráfico
        fig2 = px.pie(
            grp,
            names=col_categoria,
            values=col_pedido,
#Titulo Do Gráfico
            title="Total de Vendas Por Categoria",
            hole=0.4
        )
#Chamado Do Gráfico
        fig2.update_layout(template="plotly_dark")
        fig2.show()
```

10.CRIAÇÃO DO GRAFICO DE MAPAS

```python
#Agrupando País Por Vendas
if col_pais and col_vendas:
    grp = df.groupby(col_pais, as_index=False)[col_vendas].mean()
    if not grp.empty:
#Configurações Da Figura
        fig3 = px.choropleth(
            grp,
            locations=col_pais,
            locationmode="country names",
            color=col_vendas,
            color_continuous_scale="Viridis",
#Titulo Da Figura
            title="Média de Vendas por País"
        )
#Chamado Do Gráfico
        fig3.update_layout(template="plotly_dark")
        fig3.show()
```

11.CRIAÇÃO DO GREFICO DE BARRAS (PEDIDOS POR PAÍS)

```python
#Agrupando Paises Por Pedido
if col_pais and col_pedido:
    grp = df.groupby(col_pais, as_index=False)[col_pedido].count()
    if not grp.empty:
#Configurações Da Figura
        fig4 = px.bar(
            grp,
#Pais No Eixo X
            x=col_pais,
#Pedido No Eixo Y
            y=col_pedido,
            text_auto=True,
#Titulo Do Gráfico
            title="Pedidos por País"
        )
#Chamado Do Gráfico
        fig4.update_layout(template="plotly_dark")
        fig4.show()
```

12.CRIAÇAO DO GRÁFICO DE BARRAS (DESCONTO POR SUBCATEGORIA)

```python
#Agrupando Subcategoria Por Desconto
if col_subcategoria and col_desconto:
    grp = df.groupby(col_subcategoria, as_index=False)[col_desconto].mean()
    if not grp.empty:
#Configurações da figura
        fig5 = px.bar(
            grp,
#Subcategoria No eixo X
            x=col_subcategoria,
#Desconto No eixo Y
            y=col_desconto,
            text_auto='.2f',
#Titulo Do Gráfico
            title="Desconto Médio por Subcategoria"
          )
#Chamado Do Gráfico
        fig5.update_layout(template="plotly_dark")
        fig5.show()
```

13.CRIAÇÃO DO GRAFICO DE LINHAS

```python
#Agrupando Data Por Vendas
if col_data and col_vendas:
    grp = df.groupby(df[col_data].dt.to_period("M"))[col_vendas].sum().reset_index()
    grp[col_data] = grp[col_data].dt.to_timestamp()
#Configurações Da Figura
    fig6 = px.line(
        grp,
#Data No Eixo X
        x=col_data,
#Vendas No eixo Y
        y=col_vendas,
#Titulo Do Gráfico
        title="Evolução Mensal das Vendas"
    )
#Chamada Do Gráfico
    fig6.update_traces(mode="lines+markers")
    fig6.update_layout(template="plotly_dark")
    fig6.show()
```

RESULTADO FINAL


```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import display, HTML


arquivo = '/content/drive/MyDrive/dataset.xlsx'
df = pd.read_excel(arquivo)


df.columns = [col.strip().lower() for col in df.columns]


def encontrar_coluna(possiveis):
    for nome in possiveis:
        for c in df.columns:
            if nome in c:
                return c
    return None

col_pais = encontrar_coluna(['pais', 'country'])
col_categoria = encontrar_coluna(['categoria', 'category'])
col_subcategoria = encontrar_coluna(['sub', 'subcategoria', 'sub-category'])
col_vendas = encontrar_coluna(['venda', 'valor', 'sales', 'amount', 'total'])
col_data = encontrar_coluna(['data', 'date'])
col_desconto = encontrar_coluna(['desc', 'discount'])
col_pedido = encontrar_coluna(['pedido', 'order', 'id_pedido', 'order id'])


if col_data:
    df[col_data] = pd.to_datetime(df[col_data], errors='coerce')
    df['ano'] = df[col_data].dt.year


display(HTML("<h2 style='text-align:center;'> DASHBOARD ANÁLISE DE VENDAS GLOBAIS</h2>"))


if col_vendas:
    total = df[col_vendas].sum()
    fig1 = go.Figure(go.Indicator(
        mode="number",
        value=total,
        title={"text": "<b>Total de Vendas</b>"},
        number={"prefix": "R$", "valueformat": ",.2f"}
    ))
    fig1.update_layout(template="plotly_dark", height=250)
    fig1.show()


if col_categoria and col_pedido:
    grp = df.groupby(col_categoria, as_index=False)[col_pedido].count()
    if not grp.empty:
        fig2 = px.pie(
            grp,
            names=col_categoria,
            values=col_pedido,
            title="Total de Vendas Por Categoria",
            hole=0.4
        )
        fig2.update_layout(template="plotly_dark")
        fig2.show()


if col_pais and col_vendas:
    grp = df.groupby(col_pais, as_index=False)[col_vendas].mean()
    if not grp.empty:
        fig3 = px.choropleth(
            grp,
            locations=col_pais,
            locationmode="country names",
            color=col_vendas,
            color_continuous_scale="Viridis",
            title="Média de Vendas por País"
        )
        fig3.update_layout(template="plotly_dark")
        fig3.show()


if col_pais and col_pedido:
    grp = df.groupby(col_pais, as_index=False)[col_pedido].count()
    if not grp.empty:
        fig4 = px.bar(
            grp,
            x=col_pais,
            y=col_pedido,
            text_auto=True,
            title="Pedidos por País"
        )
        fig4.update_layout(template="plotly_dark")
        fig4.show()


if col_subcategoria and col_desconto:
    grp = df.groupby(col_subcategoria, as_index=False)[col_desconto].mean()
    if not grp.empty:
        fig5 = px.bar(
            grp,
            x=col_subcategoria,
            y=col_desconto,
            text_auto='.2f',
            title="Desconto Médio por Subcategoria"
        )
        fig5.update_layout(template="plotly_dark")
        fig5.show()


if col_data and col_vendas:
    grp = df.groupby(df[col_data].dt.to_period("M"))[col_vendas].sum().reset_index()
    grp[col_data] = grp[col_data].dt.to_timestamp()
    fig6 = px.line(
        grp,
        x=col_data,
        y=col_vendas,
        title="Evolução Mensal das Vendas"
    )
    fig6.update_traces(mode="lines+markers")
    fig6.update_layout(template="plotly_dark")
    fig6.show()
```
